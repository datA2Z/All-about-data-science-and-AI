# Simple Neural Network : Classification

In this notebook, we are creating a classifier for bank data using keras with tensorflow as backend. Based on variables available in the dataset we are predicting whether customer leaves or stay with the bank. Lets code...

```python
# Importing the libraries
import pandas as pd
import numpy as np
from sklearn import preprocessing 
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
# Path to working directory
os.chdir("F:/Deeplearning_A_Z/Artificial_Neural_Networks/Artificial_Neural_Networks")
```

```python
# Importing the dataset
data = pd.read_csv('banking.csv', header=0)
data.head()
```

```python
# Dealing with missing values in pandas
data = data.dropna()
```

```python
# We will drop the variables that we do not need.
data.drop(data.columns[[0, 3, 7, 8, 9, 10, 11, 12, 13, 15, 16, 17, 18, 19]], axis=1, inplace=True)
```

```python
# Create dummy variables
data2 = pd.get_dummies(data, columns =['job', 'marital', 'default', 'housing', 'loan', 'poutcome'])
data2.head()
```

```python
# Lets check data again
print(data2.shape)
```

We are taking 1 to 29 columns as X and first columns (index 0) as Y.

```python
X = data2.iloc[:, 1:29].values
y = data2.iloc[:, 0].values
```

```python
# Splitting the dataset into the Training set and Test set(Traning = 80% , 20% = Test)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

Transforming data in to same scale.

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Let's create ANN

```python
#import libraries
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()
# Adding the first hidden layer
classifier.add(Dense(units = 14, kernel_initializer = 'uniform', activation = 'relu', input_dim = 28))
```

```python
# Adding the second hidden layer
classifier.add(Dense(units = 14, kernel_initializer = 'uniform', activation = 'relu'))
```

```python
# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

We are runing ANN with 100 epochs but you can increase to get good results.

**Higher epochs will require more processing power. You can increase or decrease epochs based on your system configuration.**

```python
# Fitting the ANN to the Training 
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
# Lets test on test dataset
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print (cm)
```

```python
# checking accracy using K-fold cross validation
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Create funcation for K fold cross validation
def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(units = 14, kernel_initializer = 'uniform', activation = 'relu', input_dim = 28))
    classifier.add(Dense(units = 14, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
```

```python
classifier = KerasClassifier(build_fn = build_classifier, batch_size = 10, epochs = 100)
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
mean = accuracies.mean()
print (mean)
```

```python
#Let's Check Average accuracy for different models generated by K fold
print (mean)
```

```python
# Let's Check Average variance for generated models.
variance = accuracies.std()
print (variance)
```
